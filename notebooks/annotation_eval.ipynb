# Этап 3: авторазметка — оценка

Загрузите `data/labeled/labeled.csv` и `data/labeled/annotation_metrics.json`: распределение `label_auto`, гистограмма `confidence`, сравнение с эталоном `label`.

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

ROOT = Path("..").resolve()
df = pd.read_csv(ROOT / "data/labeled/labeled.csv", encoding="utf-8")
with open(ROOT / "data/labeled/annotation_metrics.json", encoding="utf-8") as f:
    m = json.load(f)
print("metrics:", m)
df["label_auto"].astype(str).value_counts().plot(kind="bar")
plt.title("label_auto")
plt.tight_layout()
plt.show()
df["confidence"].hist(bins=30)
plt.title("confidence")
plt.tight_layout()
plt.show()
```
